### ax.plot(x,y) syntex, y=x

```python
fig = plt.figure(figsize=(8, 8))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 4)
ax1.plot(x, x+5)
ax2.plot(x, 2**x)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 15))
ax[0, 0].plot(x, x+2)
ax[1, 1].plot(x, x**2)
#This is the approach we will use in the class.
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(x, x+5)
ax[1].plot(x, x**3)
#Easier than 2 lines
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.set_title('This is my Title', fontsize=25)
ax.set_xlabel('This is X Label', fontsize=20)
ax.set_ylabel('This is Y Label', fontsize=20)

ax.plot(x, x**2, label='y=x**2', color='red', linestyle=':')
ax.plot(x, x+5, label='y=x+5', color='green')
ax.legend()
ax.set_xlim(-3,12)
```
