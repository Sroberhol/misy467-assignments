There are 8 basic functions as described at http://mathonweb.com/help_ebook/html/functions_4.htm. Do the followings:

```python
import warnings
warnings.filterwarnings("ignore")

x = np.linspace(0,10,50)
z = np.linspace(-4, 4, 50)
fig = plt.figure(figsize=(15,15))
ax1 = fig.add_subplot(4, 2, 1)
ax2 = fig.add_subplot(4, 2, 2)
ax3 = fig.add_subplot(4, 2, 3)
ax4 = fig.add_subplot(4, 2, 4)
ax5 = fig.add_subplot(4, 2, 5)
ax6 = fig.add_subplot(4, 2, 6)
ax7 = fig.add_subplot(4, 2, 7)
ax8 = fig.add_subplot(4, 2, 8)
ax1.plot(x, 4*x+7, 'o', label='y=4*x+7', color='red', linestyle='solid')
ax2.plot(x, 2*x**2, label='y=2*x**2', color='blue', linestyle='dashed')
ax3.plot(z, (z**4)-15*(z**2), label='y=(z**4)-15*(z**2)', color='orange', linestyle='dashdot')
ax4.plot(z, 3*z**2 + 4*z, 'P', label='y=3*z**2 + 4*z', color='purple', linestyle='dotted')
ax5.plot(z, (2*z**2 +1)/(2*z), label='y=(2*z**2 +1)/(2*z)', color='yellow')
ax6.plot(x, 3**x, 's', label='y=3**x', color='brown')
ax7.plot(x, np.log(x), label='y=log(x)', color='green')
ax8.plot(x, np.sin(2*x), '*', label='y=sin(3*x)', color='black')
ax1.set_title('Linear', fontsize=20)
ax2.set_title('Power', fontsize=20)
ax3.set_title('Polynomial', fontsize=20)
ax4.set_title('Quadratic', fontsize=20)
ax5.set_title('Rational', fontsize=20)
ax6.set_title('Exponential', fontsize=20)
ax7.set_title('Logarithmic', fontsize=20)
ax8.set_title('Sinusoidal', fontsize=20)
ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()
ax5.legend()
ax6.legend()
ax7.legend()
ax8.legend()
ax1.set_xlabel('x')
ax1.set_ylabel('4*x+7')
ax2.set_xlabel('x')
ax2.set_ylabel('2*x**2')
ax3.set_xlabel('z')
ax3.set_ylabel('(z**4)-15*(z**2)')
ax4.set_xlabel('z')
ax4.set_ylabel('3*z**2 + 4*z')
ax5.set_xlabel('z')
ax5.set_ylabel('(2*z**2 +1)/(2*z)')
ax6.set_xlabel('x')
ax6.set_ylabel('3**x')
ax7.set_xlabel('x')
ax7.set_ylabel('log(x)')
ax8.set_xlabel('x')
ax8.set_ylabel('sin(2*x)')

fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=.4)
```
